# Numpy #
Numpy is used for numerical computing, with array & matrix support, and mathematical functions optimized to perform strongly.

## Arrays ##

```python
import numpy as np

# Create a 1D array
arr1 = np.array([1, 2, 3])
print(arr1)

# Create a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

### Built-in functions with Numpy ###

```python
# Create a 3x3 array of zeros
zeros = np.zeros((3, 3))
print(zeros)

# Create 2x2 array of ones
ones = np.ones((2, 2))
print(ones)

# Create an array with a range of values
arr_range = np.arange(0, 10, 2)
print(arr_range)

# Create an array with evenly spaced values (Ie, 0 to 1 in quarters)
linspace = np.linspace(0, 1, 5)
print(linspace)
```

### Array Manipulation ###
- Reshaping - Changing the structure of data

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped_arr = arr.reshape((2, 3))
print(reshaped_arr)
```

- Flattening - Making an array one dimensional (single row)

```python
flattened_arr = reshaped_arr.flatten()
print(flattened_arr)
```

- Transpositon - flisps columns with rows

```python
transposed_arr = reshaped_arr.T
print(transposed_arr)
```

## Mathematical Operations ##
### Element Operations ###
- Basic Operations

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

- Univeral Functions

```python
print(np.sqrt(arr1))
print(np.exp(arr1))
print(np.sin(arr1))
print(np.log(arr1))
```

### Matrix Operations ###
- Dot Product - Every row * every column

```python
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])

dot_product = np.dot(mat1, mat2) 
print(dot_product)
```

- Matrix Multiplication (Same as dot product)

```python
mat_mult = mat1 @ mat2
print(mat_mult)
```

- Determinant and Inverse

```python
det = np.linalg.det(mat1)
print(det)

inv = np.linalg.inv(mat1)
print(inv)
```

### Statistical Operations ###
- Basic Stats

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
```

- Aggregation

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])

print(np.sum(mat))
print(np.sum(mat, axis=0))  # Sum along columns
print(np.sum(mat, axis=1))  # Sum along rows
```

### Advanced Indexing and Slicing ###

- Boolean Indexing

- Fancy Indexing

### Broadcasting and Vectorization ###
- Broadcasting

```python
arr = np.array([1, 2, 3])
scalar = 2

print(arr + scalar)  # Broadcasting scalar value to each element of the array
```

- Vectorization

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])

print(mat * 2)  # Element-wise multiplication
```

### Linear Algebra ###
- Eigenvalues & Eigenvectors

```python
mat = np.array([[1, 2], [3, 4]])

eigenvalues, eigenvectors = np.linalg.eig(mat)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

- Singular Value Decompostion (SVD)

```python
U, S, V = np.linalg.svd(mat)

print("U:\n", U)
print("S:\n", S)
print("V:\n", V)
```

Pandas is a powerful data manipulation library used to load structured data into *dataframes* for use in analytics and ML.  We can also visualize the data with the data it stores.  We will demo it's most basic capabilities with the Titanic dataset.

We will load an excel file from Kaggle with the Titanic Dataset.  We will then use pandas to inspect the data.

```python
import pandas as pd

# Load the Titanic dataset from a URL
path = 'titanic3.xls'
df = pd.read_excel(path)

# Display the first few rows of the dataframe
print(df.head())

# Display basic statistics of the dataframe
print(df.describe())
```

## Data Cleaning ##
This step ensures the quality of the data by ensuring there are no redundant records, null or missing values, or inaccuracies in the data.

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values in 'age' with the median
df['age'] = df['age'].fillna(df['age'].median())

# Fill missing values in 'embarked' with the mode
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Drop the 'cabin' column as it's too sparse
df = df.drop(columns=['cabin'])
```

```python
# Check for duplicates
print(df.duplicated().sum())

# Drop duplicates if they exist
df = df.drop_duplicates()
```

```python
# Check for unique values in 'embarked'
print(df['embarked'].unique())

# Replace any incorrect values if necessary (no incorrect values found in this case)
df['embarked'] = df['embarked'].replace({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})
```

## Data Transformation ##

This step allows us to transform our data so it is better suited for analytics.  We will Create new features from the existing data, scale numerical features, and encode categorical data using one-hot encoding.

```python
# Create a new column 'FamilySize' as the sum of 'sibsp' and 'parch' + 1 (including the passenger)
df['FamilySize'] = df['sibsp'] + df['parch'] + 1

# Create a new column 'IsAlone' (1 if FamilySize is 1, else 0)
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
# Encode 'sex' as a numerical variable
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# One-hot encode 'embarked'
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
```

### Scikit-Learn ###

While this is a machine learning library, it has preprocessing and normalization tools which help us to ready our data for modeling.  We will implement a predictive model later to try and predict survival using a random forest model

```python
from sklearn.preprocessing import StandardScaler

# Scale 'age' and 'fare'
scaler = StandardScaler()
df[['age', 'fare']] = scaler.fit_transform(df[['age', 'fare']])
```

### Data Exploration & Visualization ###
With the data being clean, we can now begin our analytics of the data.  We'll pull some basic stats, create a correlation matrix, and make basic visualizations before modeling

```python
# Summary statistics
print(df.describe())

# Correlation matrix
numeric_cols = df.select_dtypes(include=['float64', 'int64']) # Only select numeric data
print(numeric_cols.corr())

```

The most basic way of making plots in python.  We will use this in conjunction with seaborn to make basic statistical graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot the distribution of 'age'
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()

# Plot the survival rate by 'sex'
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Sex')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.show()

# Plot the survival rate by 'pclass'
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

# Scikit Learn #
We will now use the sklearn library to create a basic supervised learning model.  
  
Supervised learning is 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the features and the target
features = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'body', 'FamilySize', 'IsAlone']]
target = df['survived']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

```

```python
from sklearn.tree import export_graphviz
import graphviz

# Extract one of the trees from the forest
best_tree = rf_model.estimators_[0]

# Export the tree as dot file
dot_data = export_graphviz(best_tree, out_file=None, 
                           feature_names=features.columns,  
                           class_names=['Not Survived', 'Survived'],  
                           filled=True, rounded=True,  
                           special_characters=True)  
graph = graphviz.Source(dot_data)  
graph.render("best_tree")
```
